```python
from sklearn import tree
import pandas as pd
import numpy as np
import random
```

Foi necessário substituir os valores de texto em base_model e optimizer para valores simbólicos em inteiros

```python
batch_size = list(range(4,129,16))
base_model = ['VGG16', 'VGG19', 'DenseNet121', 'ResNet50', 'InceptionV3', 'InceptionResNetV2', 'NASNetMobile', 'NASNetLarge']
image_dimension = [224,299,448,598]
optimizer = ['SGD','RMSprop','Adam','Adadelta','Adagrad','Adamax','Nadam','Ftrl']

features = [batch_size, base_model, image_dimension, optimizer] 

base_model_n = [0, 1, 2, 3, 4, 5, 6, 7]
optimizer_n = [0, 1, 2, 3, 4, 5, 6, 7]
```

Troquei o valor do loss_rate para 0,9, o valor de 0,5 estava muito baixo e gerava apenas predições valor 0

```python

df = pd.DataFrame(columns=['batch_size', 'base_model_n', 'image_dimension', 'optimizer_n', 'loss_rate'])

for batch in batch_size:
    loss_rate1 = np.random.uniform(0, 1)
    for model in base_model_n:
        loss_rate2 = loss_rate1 + np.random.uniform(0, 1)
        for image in image_dimension:
            loss_rate3 = loss_rate2 + np.random.uniform(0, 1)
            for opt in optimizer_n:
                loss_rate = loss_rate3 + np.random.uniform(0, 1)
                v = 1
                if loss_rate >= 0.9:
                    v = 0
                df.loc[len(df)] = [batch, model, image, opt, v]
df
```

```python
data = []
target = []
for index, row in df.iterrows():
    data.append([row['batch_size'], row['base_model_n'], row['image_dimension'], row['optimizer_n']])
    target.append(row['loss_rate'])
```

```python
# Randomizing 10% of the data
b_tests_dataframe = df.sample(n=int(len(df)/10))
b_tests = []
for index, row in b_tests_dataframe.iterrows():
    b_tests.append([row['batch_size'], row['base_model_n'], row['image_dimension'], row['optimizer_n']])

# Creating the tree
clf = tree.DecisionTreeClassifier(criterion="entropy")
b_tree = clf.fit(data, target)

pred = b_tree.predict(b_tests)
pred
```

```python

true_positives = 0
false_positives = 0
true_negatives = 0
false_negatives = 0

def check_test(test, pred):
    for index, row in df.iterrows():
        if (df['batch_size'] == test[0] & df['base_model_n'] == test[1] & df['image_dimension'] == test[2] & df['optimizer_n'] == test[3]):
            if df['loss_rate'] == pred:
                if pred == 1: true_positives += 1
                else: false_positives += 1
            else:
                if pred == 0: true_negatives += 1
                else: false_negatives += 1

for idx, test in enumerate(b_tests):
    check_test(test, pred[idx])

print(true_positives, false_positives, true_negatives, false_negatives)
# accuracy = 
```
