```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB
```

```python
batch_size = list(range(4,129,16))
base_model = ['VGG16', 'VGG19', 'DenseNet121', 'ResNet50', 'InceptionV3', 'InceptionResNetV2', 'NASNetMobile', 'NASNetLarge']
image_dimension = [224,299,448,598]
optimizer = ['SGD','RMSprop','Adam','Adadelta','Adagrad','Adamax','Nadam','Ftrl']

features = [batch_size, base_model, image_dimension, optimizer] 


df = pd.DataFrame(columns=['batch_size', 'base_model', 'image_dimension', 'optimizer', 'loss_rate'])

for batch in batch_size:
    loss_rate1 = np.random.uniform(0, 1)
    for model in base_model:
        loss_rate2 = loss_rate1 + np.random.uniform(0, 1)
        for image in image_dimension:
            loss_rate3 = loss_rate2 + np.random.uniform(0, 1)
            for opt in optimizer:
                loss_rate = loss_rate3 + np.random.uniform(0, 1)
                v = 1
                if loss_rate >= 1.5:
                    v = 0
                df.loc[len(df)] = [batch, model, image, opt, v]

df
```

```python
x = df[['batch_size', 'image_dimension', 'optimizer', "loss_rate"]]
x = pd.get_dummies(x, columns=['optimizer'])
y = df['base_model']

gnb = GaussianNB()
cross_val_score(gnb, x, y, cv=10).mean()
```

```python
gnb.fit(x, y)
media = pd.DataFrame(gnb.theta_, columns=x.columns, index=gnb.classes_)
desv_padrao = pd.DataFrame(gnb.sigma_, columns=x.columns, index=gnb.classes_)
```
