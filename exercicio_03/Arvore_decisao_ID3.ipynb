```python
from sklearn import tree # arvore de decisao
from sklearn.datasets import load_iris #dataset
from sklearn.metrics import confusion_matrix
import numpy as np
import random
```

```python
# carregamos o dataset 
iris = load_iris()
# separamos as features e os targets
X = iris.data
y = iris.target
test_x = []
test_y = []
for i in range(15):
    j = random.randint(0,len(X)-1)
    test_x.append(X[j])
    test_y.append(y[j])
    X = np.delete(X,j,0)
    y = np.delete(y,j,0)
```

```python
print(X.shape)
print(y.shape)
```

```python
# Definimos a árvore de decisão com o critério de entropia
clf = tree.DecisionTreeClassifier(criterion="entropy")
```

```python
irisTree = clf.fit(X, y)
```

```python
y_pred = irisTree.predict(test_x)
```

```python
print(y_pred)
print(test_y)
```

```python
confusion_matrix(test_y, y_pred)
```
