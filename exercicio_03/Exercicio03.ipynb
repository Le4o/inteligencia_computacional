Considere os dados do exercício 02. Sua tarefa é construir uma árvore de decisão para classificar conjunto de parâmetros (batch_size, base_model, image_dimension, optimizer) são úteis ou não.

Para isso considere:
  - Se loss_rate < 0.5 significa que o conjunto de parâmetros são úteis (classe positiva)
  - Se o loss loss_rate >= 0.5 significa que o conjunto de parâmetros são não úteis (classe negativa)

Você deve separar aleatoriamente 10% dos dados para ser sua base de teste e deve calcular a partir da matriz de confusão a: acurácia, precisão e recall.

```python
batch_size = list(range(4,129,16))
base_model = ['VGG16', 'VGG19', 'DenseNet121', 'ResNet50', 'InceptionV3', 'InceptionResNetV2', 'NASNetMobile', 'NASNetLarge']
image_dimension = [224,299,448,598]
optimizer = ['SGD','RMSprop','Adam','Adadelta','Adagrad','Adamax','Nadam','Ftrl']

caracteristicas = [batch_size, base_model, image_dimension, optimizer] 

[4, 'VGG19', 299, 'Adam']
```

```python
import pandas as pd
import numpy as np
import random
df = pd.DataFrame(columns=['batch_size', 'base_model', 'image_dimension', 'optimizer', 'loss_rate'])

for batch in batch_size:
    loss_rate1 = np.random.uniform(0, 1)
    for model in base_model:
        loss_rate2 = loss_rate1 + np.random.uniform(0, 1)
        for image in image_dimension:
            loss_rate3 = loss_rate2 + np.random.uniform(0, 1)
            for opt in optimizer:
                loss_rate = loss_rate3 + np.random.uniform(0, 1)
                v = 1
                if loss_rate >= 0.5:
                    v = 0
                df.loc[len(df)] = [batch, model, image, opt, v]

df
```
